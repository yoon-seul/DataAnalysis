# Loading Data

```python
data=pd.read_csv("marketing_campaign.csv",sep='\t')
```

since last update is 5 months ago, I will set 2022 as standard of counting age.<br>

# Data Visualization

Visualize some columns of the dataset.

```python
plt.figure(figsize=(10,75))

plt.subplot(15,2,1)
sns.distplot(data['Age'],color='navy')

plt.subplot(15,2,2)
education=data['Education'].value_counts()
plt.pie(education,autopct='%.1f%%',labels=[education.index[0],education.index[1],education.index[2],education.index[3],education.index[4]])

plt.subplot(15,2,3)
marry=sns.countplot(data['Marital_Status'])
marry.set_xticklabels(marry.get_xticklabels(),rotation=45)
for p in marry.patches:
    height = p.get_height()
    marry.text(p.get_x() + p.get_width() / 2., height + 5, height, ha = 'center', size = 9)
marry

plt.subplot(15,2,4)
sns.distplot(data['Income'],color='navy')

plt.subplot(15,2,5)
kid=data['Kidhome'].value_counts()
plt.pie(kid,autopct='%.1f%%',labels=[kid.index[0],kid.index[1],kid.index[2]])

plt.subplot(15,2,6)
teen=data['Teenhome'].value_counts()
plt.pie(teen,autopct='%.1f%%',labels=[teen.index[0],teen.index[1],teen.index[2]])
```

Seeing Pearson correlation between columns.

```python
plt.figure(figsize=(14,7))
mask=np.zeros_like(data.corr(),dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

plt.subplot(1,2,1)
sns.heatmap(new_data.corr()>0.4,mask=mask,annot=True)
plt.subplot(1,2,2)
sns.heatmap(new_data.corr()<-0.4,mask=mask,annot=True)
```

```python
correlation=[]
for n in new_data.corr().columns:
    for m in new_data.corr().columns:
        if ((new_data.corr()>0.4) & (new_data.corr() < 1))[n][m]==True:
            correlation.append((n,m))
        if (new_data.corr()<-0.4)[n][m]==True:
            correlation.append((n,m))
```

```python
correlation
```

It seems there are a lot of correlation between the products and income.<br>
So, let's look them.

```python
products=new_data.loc[:,['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']]
```

```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
sns.scatterplot(data=products,x='Income',y='MntWines',color='navy')
plt.subplot(2,2,2)
sns.scatterplot(data=products,x='Income',y='MntFruits',color='navy')
plt.subplot(2,2,3)
sns.scatterplot(data=products,x='Income',y='MntMeatProducts',color='navy')
plt.subplot(2,2,4)
sns.scatterplot(data=products,x='Income',y='MntFishProducts',color='navy')
```
